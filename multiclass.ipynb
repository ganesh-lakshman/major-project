```python
import numpy as np
import keras
from sklearn import datasets
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
```

```python
n_pts = 500
# centers of clusters(blobs)
centers = [[-1, 1], [-1, -1], [1, -1]]
# try with 5 cluster [1,1] [0,0]
# std deviation, if it is too small data points will be densely packed large vice versa
X, y = datasets.make_blobs(n_samples=n_pts, random_state = 123, centers=centers, cluster_std=0.4)

# X,y
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1]) # [-1, 1]
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.scatter(X[y==2, 0], X[y==2, 1])
```

```python
# hot encoding
# print(y)
y_cat = to_categorical(y, 3)
# y_cat
```

```python
# NN single ip layer single op layer
model = Sequential()
model.add(Dense(units = 3, input_shape=(2,), activation='softmax'))
model.compile(Adam(learning_rate=0.1),loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=X, y=y_cat, verbose=1, batch_size = 50, epochs=100)
```

```python
def plot_decision_boundary(X, y_cat, model):
    x_span = np.linspace(min(X[:, 0]) - 0.25, max(X[:, 0]) + 0.25, 50)
    y_span = np.linspace(min(X[:, 1]) - 0.25, max(X[:, 1]) + 0.25, 50)
    # print(x_span)
    xx, yy = np.meshgrid(x_span, y_span)
    xx_, yy_ = xx.ravel(), yy.ravel()
    grid = np.c_[xx_, yy_]
    # print(grid)
    pred_func = np.argmax(model.predict(grid), axis=-1)
    z = pred_func.reshape(xx.shape)
    plt.contourf(xx, yy, z)
```

```python
plot_decision_boundary(X, y_cat, model)
plt.scatter(X[y==0, 0], X[y==0, 1]) # [-1, 1]
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.scatter(X[y==2, 0], X[y==2, 1])
```

```python
plot_decision_boundary(X, y_cat, model)
plt.scatter(X[y==0, 0], X[y==0, 1]) # [-1, 1]
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.scatter(X[y==2, 0], X[y==2, 1])

# x=0.5
# y=0.5
x=0.5
y=-1
point = np.array([[x,y]])
prediction = np.argmax(model.predict(point), axis=-1)
plt.plot([x],[y],marker='o', markersize=10,color='r')
print("prediction is: ", prediction)
```
